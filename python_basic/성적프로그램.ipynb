```python
import numpy as np
import pandas as pd
```

```python
kor_score = [70, 79, 80, 90, 80] 
math_score = [42, 80, 30, 50, 90] 
eng_score = [53, 77, 50, 70, 55]
```

```python
pd.DataFrame({'kor':kor_score, 'math':math_score,'eng':eng_score})
# (계산) 데이터 프레임을 활용하여 표로 만들어 놓기 
```

```python
score_df=pd.DataFrame({'kor':kor_score, 'math':math_score,'eng':eng_score})
```

```python
score_df
```

```python
score_df.describe()
# 과목별 평균
```

```python
score_df.T.describe()
# 학생별 평균
```

```python
# 국영수 평균이 90 이상이면 A, 80이상이면 B, 70이상이면 C, 그외에는 D로 표시
# pd.select() 활용
```

```python
stu_mean = score_df.T.describe()
```

```python
stu_mean.index
# 데이터프레임에서 mean(평균)만 추출하기 위해 index 출력
```

```python
stu_mean.loc['mean']
# 인덱스에서 불러올때는 위치를 알려주는 loc['인덱스명']
```

```python
pd.DataFrame(stu_mean.loc['mean'])
# 시리즈를 다시 데이터 프레임으로 변환
```

```python
stu_score = pd.DataFrame(stu_mean.loc['mean'])
```

```python
# pd.select() 함수

# 조건 : 조건 변수를 정하고 리스트로 정리
```

```python
score_lb = [
    stu_mean.loc['mean'] > 80,
    stu_mean.loc['mean'] > 70,
    stu_mean.loc['mean'] > 60
]
```

```python
# 조건 결과 : 위 조건에 따른 결과를 순서대로 리스트로 정리
```

```python
result = ['A','B','C']
```

```python
np.select(score_lb, result, default='-')
```
