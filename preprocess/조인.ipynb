```python
df_left=pd.DataFrame({'Name':['Harry Potter','David Baker','John Smith','Juan Martinez','Jane Connor'],
                      'Gender':['Male','Male','Male','Male','Female'],'Age':[23,31,22,36,30]})
```

```python
df_right=pd.DataFrame({'Name':['John Smith','Alex Du Bois','Joanne Rowling','Jane Connor'],
                      'Position':['Intern','Team Lead','Manager','Manager'],'Wage':[25000,75000,90000,70000]})
```

```python
df_left
```

```python
df_right
```

```python
df_left.columns , df_right.columns
```

```python
pd.merge(df_left,df_right,on='Name')
```

```python
pd.merge(df_left,df_right,on='Name', how='inner') #innerjoin
```

```python
pd.merge(df_left,df_right,on='Name', how='left')
```

```python
pd.merge(df_left,df_right,on='Name', how='right')
```

```python
pd.merge(df_left,df_right,on='Name', how='outer')
```
