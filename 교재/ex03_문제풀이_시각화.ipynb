```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#  문제 1
```

```python
time_of_day =[12,13,14,15,16,17,18]
exercise_time =[50,40,35,30,20,25,55]
```

```python
plt.plot(time_of_day,exercise_time, marker="o")
plt.title('Exercise Time by Time of Day')
plt.xlabel('Time of Day(Hour)')
plt.ylabel('Exercise Time (Minutes)')
plt.grid()
plt.show()
```

```python
exercise = ['Pilates', 'Strength Training', 'Yoga']
calories_burned =[200,450,100]
```

```python
# 문제 3
```

```python
equipment = ['Bike','Weights','Dumbbells']      #x
usage_count = [150, 180, 80]                    #y1
total_calories = [5000, 6000, 3500]             #y2
```

```python
fig, ax1 = plt.subplots()
ax1.plot(equipment, usage_count, marker='o', color='#9b59b6', label='Usage Count', linewidth=5)
ax1.set_xlabel('Exercise Equipment')
ax1.set_ylabel('Usage Count', color='#9b59b6')
ax1.tick_params(axis='y', labelcolor='#9b59b6')
ax1.legend(loc='upper right')

ax2 = ax1.twinx()
ax2.bar(equipment, total_calories, color='#f5b7b1', alpha=0.7, label='Total Calories Burned')
ax2.set_ylabel('Total Calories Burned', color='#f5b7b1')
ax2.tick_params(axis='y', labelcolor='#f5b7b1')
ax2.legend(loc='upper right', bbox_to_anchor=(1,0.9))

plt.suptitle('Exercise Equipment Usage and Calories Burned Analysis')
plt.tight_layout()
plt.show()
```
